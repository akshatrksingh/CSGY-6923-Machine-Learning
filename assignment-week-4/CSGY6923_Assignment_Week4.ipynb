```python
import pandas as pd
```

```python
df1 = pd.read_csv("../assignment-week-3/kddcup99_csv.csv")
df1.head()
```

```python
df1["label"].value_counts()
```

```python
df2 = df1[df1["label"] != "normal"]
df2
```

```python
df3 = df2.copy()
df3.loc[df3["label"].isin(["back", "land", "neptune", "pod", "smurf", "teardrop"]), "label"] = 1  # dos - class 1
df3.loc[df3["label"] != 1, "label"] = 0 # non dos - class 0
df3 = df3.reset_index(drop=True)
df3
```

```python
df3["label"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
categorical_columns = df3.select_dtypes(include=['object']).columns.tolist()[:-1]
one_hot_encoded = encoder.fit_transform(df3[categorical_columns])
one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
one_hot_encoded_df
```

```python
categorical_columns
```

```python
df4 = pd.concat([one_hot_encoded_df, df3], axis=1).drop(columns=categorical_columns)
df4["label"] = df4["label"].astype(int)
df4
```

```python
df4["label"].value_counts()
```

```python
X = df4.iloc[:,:-1]
y = df4.iloc[:,-1]
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

kernels = ['linear', 'rbf', 'poly', 'sigmoid']

for kernel in kernels:
    print(f"\nTraining SVM with {kernel} kernel")
    
    svm_model = SVC(kernel=kernel, random_state=42)
    svm_model.fit(X_train, y_train)
    
    y_pred = svm_model.predict(X_test)
    
    print(f"Confusion Matrix for {kernel} kernel:")
    print(confusion_matrix(y_test, y_pred))
    
    print(f"Classification Report for {kernel} kernel:")
    print(classification_report(y_test, y_pred))
```
