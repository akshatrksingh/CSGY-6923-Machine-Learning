```python
df3 = df2.copy()
df3.loc[df3["label"].isin(["ipsweep", "portsweep", "nmap", "satan"]), "label"] = 1  # probe - class 1
df3.loc[df3["label"] != 1, "label"] = 0 # not probe - class 0
df3 = df3.reset_index(drop=True)
df3
```

```python
df3["label"].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
categorical_columns = df3.select_dtypes(include=['object']).columns.tolist()[:-1]
one_hot_encoded = encoder.fit_transform(df3[categorical_columns])
one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))
one_hot_encoded_df
```

```python
categorical_columns
```

```python
df4 = pd.concat([one_hot_encoded_df, df3], axis=1).drop(columns=categorical_columns)
df4["label"] = df4["label"].astype(int)
df4
```

```python
df4["label"].value_counts()
```

```python
X = df4.iloc[:,:-1]
y = df4.iloc[:,-1]
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=5, penalty=None)
model.fit(X_train_scaled, y_train)
```

```python
train_accuracy = model.score(X_train_scaled, y_train)
test_accuracy = model.score(X_test_scaled, y_test)

print(f"Training Accuracy: {round(train_accuracy, 5)}")
print(f"Test Accuracy: {round(test_accuracy, 5)}")
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_scaled)
report = classification_report(y_test, y_pred)

print(report)
```

```python
n_iterations = model.n_iter_

print(f"Number of iterations: {n_iterations}")
```

Training accuracy is 99.989 % and test accuracy is 99.991 %. Given that both the training and testing accuracies are almost identical and very high, overfitting does not appear to be an issue in this case. It took 19 iterations to converge

```python
weights = model.coef_[0]
abs_weights = [abs(round(weight, 4)) for weight in weights]
```

```python
features = df4.columns[:-1] 
features
```

```python
sorted_items = sorted(zip(features, abs_weights), key=lambda item: item[1], reverse=True)
feature_weight_dict = {feature: abs_weight for feature, abs_weight in sorted_items}
feature_weight_dict
```

The features having the largest weights in absolute value are wrong_fragment (number of wrong fragments), dst_host_diff_srv_rate () and so on. Yes they make sense because a high value of number of wrong fragments obviously suggest suspicious activity.

```python
model_l1 = LogisticRegression(random_state=5, penalty="l1", solver="liblinear")
model_l1.fit(X_train_scaled, y_train)
```

```python
train_accuracy = model_l1.score(X_train_scaled, y_train)
test_accuracy = model_l1.score(X_test_scaled, y_test)

print(f"Training Accuracy for l1: {round(train_accuracy, 5)}")
print(f"Test Accuracy for l1: {round(test_accuracy, 5)}")
```

```python
n_iterations = model_l1.n_iter_

print(f"Number of iterations for l1: {n_iterations}")
```

```python
model_l2 = LogisticRegression(random_state=5, penalty="l2", solver="liblinear")
model_l2.fit(X_train_scaled, y_train)
```

```python
train_accuracy = model_l2.score(X_train_scaled, y_train)
test_accuracy = model_l2.score(X_test_scaled, y_test)

print(f"Training Accuracy for l2: {round(train_accuracy, 5)}")
print(f"Test Accuracy for l2: {round(test_accuracy, 5)}")
```

```python
n_iterations = model_l2.n_iter_

print(f"Number of iterations for l2: {n_iterations}")
```

The accuracies for both L1 and L2 regularized models are similar. This outcome can be attributed to the fact that before applying regularization, our initial model training did not exhibit signs of overfitting. Both models were able to generalize well to the validation data, indicating that the complexity of the models was appropriate for the dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

weights_l1 = model_l1.coef_[0]
weights_l2 = model_l2.coef_[0]

feature_names = np.array(features) 

plt.figure(figsize=(24, 24)) 

plt.subplot(2, 1, 1)
plt.barh(feature_names, weights_l1, color='skyblue')
plt.title('Feature Weights (L1 Regularization)', fontsize=14)
plt.xlabel('Weight', fontsize=12)
plt.axvline(0, color='black', linestyle='--')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.subplot(2, 1, 2)
plt.barh(feature_names, weights_l2, color='salmon')
plt.title('Feature Weights (L2 Regularization)', fontsize=14)
plt.xlabel('Weight', fontsize=12)
plt.axvline(0, color='black', linestyle='--')
plt.xticks(fontsize=10)
plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
C_values = [0.001, 0.01, 0.1, 1, 10]
train_accuracies = []
test_accuracies = []

for C in C_values:
    model = LogisticRegression(random_state=5, penalty="l2", C=C, solver="liblinear")
    model.fit(X_train_scaled, y_train)
    
    y_train_pred = model.predict(X_train_scaled)
    y_test_pred = model.predict(X_test_scaled)
    
    train_accuracy = model.score(X_train_scaled, y_train)
    test_accuracy = model.score(X_test_scaled, y_test)
    
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)
    
for C, train_acc, test_acc in zip(C_values, train_accuracies, test_accuracies):
    print(f"C = {C:.3f} | Train Accuracy: {train_acc:.4f} | Test Accuracy: {test_acc:.4f}")
```

As C increases (indicating weaker regularization), the model tends to fit the training data more closely, often resulting in higher accuracy. In these results, the higher values yield near-perfect accuracies, showing effective learning, but this also raises the risk of overfitting. Conversely, smaller C values lead to stronger regularization, which can help prevent overfitting but may result in lower accuracy due to underfitting. Finding the right balance for C is crucial to ensure that the model generalizes well to unseen data.
